# Construct one Feature Table
## By merging scaled feature vectors

Create a feature table-:

```python
features = []

for items in mfcc_cel_scaled:
    features = features + [items]
for items in mfcc_cla_scaled:
    features = features + [items]
for items in mfcc_flu_scaled:
    features = features + [items]
for items in mfcc_gac_scaled:
    features = features + [items]
for items in mfcc_gel_scaled:
    features = features + [items]
for items in mfcc_org_scaled:
    features = features + [items]
for items in mfcc_pia_scaled:
    features = features + [items]
for items in mfcc_sax_scaled:
    features = features + [items]
for items in mfcc_tru_scaled:
    features = features + [items]
for items in mfcc_vio_scaled:
    features = features + [items]
for items in mfcc_voi_scaled:
    features = features + [items]
 
features = numpy.vstack(features)
features.shape
```

Construct a vector of labels, where 0 refers to the cel audio file, and 1 refers to cla audio file and so one.

```python
cel_labels = numpy.full(len(mfcc_cel_scaled[0])*len(mfcc_cel_scaled),0)
cla_labels = numpy.full(len(mfcc_cla_scaled[0])*len(mfcc_cla_scaled),1)
flu_labels = numpy.full(len(mfcc_flu_scaled[0])*len(mfcc_flu_scaled),2)
gac_labels = numpy.full(len(mfcc_gac_scaled[0])*len(mfcc_gac_scaled),3)
gel_labels = numpy.full(len(mfcc_gel_scaled[0])*len(mfcc_gel_scaled),4)
org_labels = numpy.full(len(mfcc_org_scaled[0])*len(mfcc_org_scaled),5)
pia_labels = numpy.full(len(mfcc_pia_scaled[0])*len(mfcc_pia_scaled),6)
sax_labels = numpy.full(len(mfcc_sax_scaled[0])*len(mfcc_sax_scaled),7)
tru_labels = numpy.full(len(mfcc_tru_scaled[0])*len(mfcc_tru_scaled),8)
vio_labels = numpy.full(len(mfcc_vio_scaled[0])*len(mfcc_vio_scaled),9)
voi_labels = numpy.full(len(mfcc_voi_scaled[0])*len(mfcc_voi_scaled),10)

labels = numpy.concatenate((cel_labels,cla_labels,flu_labels,gac_labels,gel_labels,org_labels,pia_labels,sax_labels,tru_labels,vio_labels,voi_labels))
labels.shape
```

# Train a Classifier

### Multilayer Perceptron (MLP)

```python
from sklearn.neural_network import MLPClassifier
model =  MLPClassifier(alpha=1)
```

Train the classifier-:

```python
model.fit(features,labels)
```

# Run the classifier

Let's test the classifier on x_cel (loaded in the beginning) which was not included in the training set

```python
mfcc_cel_test = librosa.feature.mfcc(x_cel, sr=sr_cel, n_mfcc=n_mfcc).T
mfcc_cla_test = librosa.feature.mfcc(x_cla, sr=sr_cla, n_mfcc=n_mfcc).T
mfcc_flu_test = librosa.feature.mfcc(x_flu, sr=sr_flu, n_mfcc=n_mfcc).T
mfcc_gac_test = librosa.feature.mfcc(x_gac, sr=sr_gac, n_mfcc=n_mfcc).T
mfcc_gel_test = librosa.feature.mfcc(x_gel, sr=sr_gel, n_mfcc=n_mfcc).T
mfcc_org_test = librosa.feature.mfcc(x_org, sr=sr_org, n_mfcc=n_mfcc).T
mfcc_pia_test = librosa.feature.mfcc(x_pia, sr=sr_pia, n_mfcc=n_mfcc).T
mfcc_sax_test = librosa.feature.mfcc(x_sax, sr=sr_sax, n_mfcc=n_mfcc).T
mfcc_tru_test = librosa.feature.mfcc(x_tru, sr=sr_tru, n_mfcc=n_mfcc).T
mfcc_vio_test = librosa.feature.mfcc(x_vio, sr=sr_vio, n_mfcc=n_mfcc).T
mfcc_voi_test = librosa.feature.mfcc(x_voi, sr=sr_voi, n_mfcc=n_mfcc).T
```

Scale the MFCCs using the previous scaler:

```python
mfcc_cel_test_scaled = scaler.transform(mfcc_cel_test)
mfcc_cla_test_scaled = scaler.transform(mfcc_cla_test)
mfcc_flu_test_scaled = scaler.transform(mfcc_flu_test)
mfcc_gac_test_scaled = scaler.transform(mfcc_gac_test)
mfcc_gel_test_scaled = scaler.transform(mfcc_gel_test)
mfcc_org_test_scaled = scaler.transform(mfcc_org_test)
mfcc_pia_test_scaled = scaler.transform(mfcc_pia_test)
mfcc_sax_test_scaled = scaler.transform(mfcc_sax_test)
mfcc_tru_test_scaled = scaler.transform(mfcc_tru_test)
mfcc_vio_test_scaled = scaler.transform(mfcc_vio_test)
mfcc_voi_test_scaled = scaler.transform(mfcc_voi_test)
```

```python
features_test = numpy.vstack((mfcc_cel_test_scaled,mfcc_cla_test,mfcc_flu_test_scaled,mfcc_gac_test_scaled,mfcc_gel_test_scaled,mfcc_org_test_scaled,mfcc_pia_test_scaled,mfcc_sax_test_scaled,mfcc_tru_test_scaled,mfcc_vio_test_scaled,mfcc_voi_test_scaled))
features_test.shape
```

```python
tcel_labels = numpy.full(len(mfcc_cel_test_scaled),0)
tcla_labels = numpy.full(len(mfcc_cla_test_scaled),1)
tflu_labels = numpy.full(len(mfcc_flu_test_scaled),2)
tgac_labels = numpy.full(len(mfcc_gac_test_scaled),3)
tgel_labels = numpy.full(len(mfcc_gel_test_scaled),4)
torg_labels = numpy.full(len(mfcc_org_test_scaled),5)
tpia_labels = numpy.full(len(mfcc_pia_test_scaled),6)
tsax_labels = numpy.full(len(mfcc_sax_test_scaled),7)
ttru_labels = numpy.full(len(mfcc_tru_test_scaled),8)
tvio_labels = numpy.full(len(mfcc_vio_test_scaled),9)
tvoi_labels = numpy.full(len(mfcc_voi_test_scaled),10)

labels_test = numpy.concatenate((tcel_labels,tcla_labels,tflu_labels,tgac_labels,tgel_labels,torg_labels,tpia_labels,tsax_labels,ttru_labels,tvio_labels,tvoi_labels))
labels_test.shape
```

Compute the predicted labels-:

```python
predicted_labels = model.predict(features_test)
predicted_labels
```

Compute the accuracy score of the classifier on the test data

```python
score = model.score(features_test,labels_test)
```
